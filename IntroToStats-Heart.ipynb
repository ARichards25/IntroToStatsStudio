## Section 3: Playing with our data

Lets start by breaking our data into two.  

```python
#using catagorical data and groupby
df_death_event = df.groupby('DEATH_EVENT')
df_died = df_death_event.get_group(1)
   

df_died.corr()
```

```python
for column in df_died:
        plt.figure(figsize=(17,1))
        sns.boxplot(data=df_died)
```

```python
df_died.hist(figsize = (20,20))

plt.show()
```

```python
# Graph heatmap to visualize correlations(correlation matrix)
fig, ax = plt.subplots(figsize = (14,10))
ax = sns.heatmap(df_died.corr(), annot=True, fmt='.2f')
plt.show()
```
